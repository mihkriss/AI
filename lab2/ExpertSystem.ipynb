```python
class Planets(KnowledgeEngine):
    @Rule(AND(Fact("Есть сосуды"), Fact("Есть органы")))
    def Higher(self):
        self.declare(Fact("Высшие"))

    @Rule(
        OR(
            Fact("Одноклеточные"),
            Fact("Нет половых органов"),
            Fact("Есть таллом и слоевище"),
        )
    )
    def Lower(self):
        self.declare(Fact("Низшие"))

    @Rule(AND(Fact("Высшие"), Fact("Размножение спорами")))
    def Spore(self):
        self.declare(Fact("Споровые"))

    @Rule(AND(Fact("Высшие"), Fact("Размножение семенами")))
    def Seeds(self):
        self.declare(Fact("Семенные"))

    @Rule(
        AND(
            Fact("Споровые"),
            Fact("Побег: весенний и летний"),
            Fact("Зеленое растение"),
            Fact("Заросток"),
            Fact("Размножение корневищами"),
            Fact("Корневище"),
            Fact("Мелкие"),
            Fact("Чешуевидные"),
            Fact("Прямостоящий"),
            Fact("Жесткий"),
        )
    )
    def Horsetails(self):
        self.declare(Fact("Хвощеобразные"))

    @Rule(
        AND(
            Fact("Споровые"),
            Fact("Зеленое растение"),
            Fact("Заросток"),
            Fact("Размножение побегами"),
            Fact("Придаточные"),
            Fact("Мелкие"),
            Fact("Различной формы"),
            OR(Fact("Прямостоящий"), Fact("Cтелющийся")),
        )
    )
    def Lycosformes(self):
        self.declare(Fact("Плаунообразные"))

    @Rule(
        AND(
            Fact("Споровые"),
            Fact("Нить, коробочка со спорами"),
            Fact("Листостебельное растение"),
            Fact("Нет корней"),
        )
    )
    def Bryophytes(self):
        self.declare(Fact("Мохообразные"))

    @Rule(
        AND(
            Fact("Споровые"),
            Fact("Зеленое растение"),
            Fact("Заросток"),
            Fact("Размножение почками на корневище"),
            Fact("Придаточные"),
            Fact("Вайи"),
            Fact("Короткий"),
            Fact("Жесткий"),
        )
    )
    def Ferns(self):
        self.declare(Fact("Папоротникообразные"))

    @Rule(AND(Fact("Хвощеобразные"), Fact("Растет на лугу")))
    def Horsetail(self):
        self.declare(Fact(planet="Хвощ луговой"))

    @Rule(AND(Fact("Хвощеобразные"), Fact("Растет на болоте")))
    def Marsh_horsetail(self):
        self.declare(Fact(planet="Хвощ болотный"))

    @Rule(AND(Fact("Плаунообразные"), Fact("Похожи на мхи")))
    def Lycopsidae(self):
        self.declare(Fact("Плауновые"))

    @Rule(AND(Fact("Плаунообразные"), Fact("Листьями напоминают злаки")))
    def Half_shell(self):
        self.declare(Fact("Полушниковые"))

    @Rule(AND(Fact("Плаунообразные"), Fact("Побеги напоминают вайи")))
    def Selaginella(self):
        self.declare(Fact("Селагинелловые"))

    @Rule(
        AND(
            Fact("Плауновые"),
            Fact("Листья прижаты к стеблю"),
            Fact("Ползучие побеги"),
            Fact("Спорангии в стробилах"),
        )
    )
    def Double_row(self):
        self.declare(Fact(planet="Двурядник"))

    @Rule(
        AND(Fact("Плауновые"), Fact("Ползучие побеги"), Fact("Спорангии в стробилах"))
    )
    def Club_moss(self):
        self.declare(Fact(planet="Плаун"))

    @Rule(
        AND(
            Fact("Плауновые"),
            Fact("Ползучие побеги"),
            Fact("Спорангии на неспециализированных побегах"),
        )
    )
    def Baranets(self):
        self.declare(Fact(planet="Баранец"))

    @Rule(Fact("Полушниковые"))
    def Polushnik(self):
        self.declare(Fact(planet="Полушник"))

    @Rule(AND(Fact("Селагинелловые"), Fact("Мелкие листья до 5 мм")))
    def Plaunok_scaly(self):
        self.declare(Fact(planet="Плаунок чешуелистный"))

    @Rule(AND(Fact("Селагинелловые"), Fact("Внутри листья рыжие")))
    def Mollendorff_float(self):
        self.declare(Fact(planet="Плаунок Мёллендорфа"))

    @Rule(
        AND(
            Fact("Мохообразные"),
            Fact("Листья на боковых веточках"),
            Fact("Рассеивание спор при растрескивании коробочки"),
        )
    )
    def Famsphagnum_mossesily(self):
        self.declare(Fact("Сфагновые мхи"))

    @Rule(
        AND(
            Fact("Мохообразные"),
            Fact("Листья на главном стебле"),
            Fact("Рассеивание спор при раскрытии зубчиков коробочки"),
        )
    )
    def Green_mosses(self):
        self.declare(Fact("Зеленые мхи"))

    @Rule(AND(Fact("Зеленые мхи"), Fact("Листья собраны в розетку")))
    def Brium(self):
        self.declare(Fact(planet="Бриум"))

    @Rule(AND(Fact("Зеленые мхи"), Fact("Листья на стебле и ветках")))
    def Kratoneur(self):
        self.declare(Fact(planet="Кратонеур"))

    @Rule(AND(Fact("Сфагновые мхи"), Fact("Образует торф")))
    def Sphagnum(self):
        self.declare(Fact(planet="Сфагнум"))

    @Rule(
        AND(
            Fact("Папоротникообразные"),
            Fact("Спорангии со стенкой толщиной в 1 клетку"),
        )
    )
    def Ferns(self):
        self.declare(Fact(family="Папоротниковые"))

    @Rule(
        AND(Fact("Папоротникообразные"), Fact("Лист и апекс побега направлены вверх"))
    )
    def Marattiaceae(self):
        self.declare(Fact(family="Мараттиевые"))

    @Rule(AND(Fact("Папоротниковые"), Fact("Листья кожистые"), Fact("Вайи перистые")))
    def Chistoust(self):
        self.declare(Fact(planet="Чистоуст"))

    @Rule(AND(Fact("Папоротниковые"), Fact("Листья кожистые")))
    def Hymenophyllum(self):
        self.declare(Fact(planet="Гименофиллум"))

    @Rule(AND(Fact("Папоротниковые"), Fact("Листья как у клевера")))
    def Marsilia(self):
        self.declare(Fact(planet="Марсилия"))

    @Rule(AND(Fact("Мараттиевые"), Fact("Листья до 5-6 м")))
    def Angiopteris(self):
        self.declare(Fact(planet="Ангиоптерис"))

    @Rule(AND(Fact("Мараттиевые"), Fact("Листья до 1-2 м")))
    def Marattia(self):
        self.declare(Fact(planet="Мараттия"))

    # Семенные
    @Rule(
        Fact("Семенные"),
        Fact("Опыление - ветром"),
        Fact("Семязачатки голые лежат на чешуе шишке"),
    )
    def Gymnosperms(self):
        self.declare(Fact("Голосеменные"))

    # Голосеменные
    @Rule(
        Fact("Голосеменные"),
        Fact("Растут в засушливых обл."),
        Fact("Семя окружено околоплодником"),
    )
    def Oppressive(self):
        self.declare(Fact("Гнетовые"))

    @Rule(
        Fact("Голосеменные"),
        Fact("Растут в жарких влажных обл."),
        Fact("Однополые стробилы"),
    )
    def Cycads(self):
        self.declare(Fact("Саговниковые"))

    @Rule(Fact("Голосеменные"), Fact("Растут в тайге"), Fact("Есть шишки"))
    def Conifers(self):
        self.declare(Fact("Хвойные"))

    # Гнетовые
    @Rule(Fact("Гнетовые"), Fact("Длинные тонкие побеги"), Fact("Кустарникм"))
    def Ephedra(self):
        self.declare(Fact(planet="Эфедра"))

    @Rule(Fact("Гнетовые"), Fact("Крупные листья"), Fact("Кустарникм"))
    def Gnetum_gnemon(self):
        self.declare(Fact(planet="Гнетум гнемон"))

    @Rule(Fact("Гнетовые"), Fact("Живёт до 2000 лет"), Fact("Живёт до 2000 лет"))
    def Velvichia(self):
        self.declare(Fact(planet="Вельвичия"))

    # Саговниковые
    @Rule(Fact("Саговниковые"))
    def Cycad(self):
        self.declare(Fact(planet="Саговник"))

    # Хвойные
    @Rule(Fact("Хвойные"), Fact("Короткие иголки"), Fact("Коническая крона"))
    def Spruce(self):
        self.declare(Fact(planet="Ель"))

    @Rule(Fact("Хвойные"), Fact("Очень высокое"), Fact("Коническая крона"))
    def Sequoia(self):
        self.declare(Fact(planet="Секвойа"))

    @Rule(Fact("Хвойные"), Fact("Хвоя однолетняя"))
    def Larch(self):
        self.declare(Fact(planet="Лиственница"))

    @Rule(Fact("Хвойные"), Fact("Съедобные орешки"))
    def Cedar(self):
        self.declare(Fact(planet="Кедр"))

    @Rule(Fact("Хвойные"), Fact("Длинные иголки"), Fact("Золотистый ствол"))
    def Pine(self):
        self.declare(Fact(planet="Сосна"))

    @Rule(Fact("Хвойные"), Fact("Пирамидальная кронаи"))
    def Cypressr(self):
        self.declare(Fact(planet="Кипарис"))

    @Rule(Fact("Хвойные"), Fact("Неприятный запах"), Fact("Листья линейные"))
    def Torreya(self):
        self.declare(Fact(planet="Торрея"))

    # Покрытосеменные

    # Низшие
    @Rule(
        AND(
            Fact("Низшие"),
            Fact("Запасное вещество - крахмал"),
            Fact("Запасное вещество - масла"),
        )
    )
    def Dinophyte_algae(self):
        self.declare(Fact(planet="Динофитовые водоросли"))

    @Rule(
        AND(Fact("Низшие"), Fact("Многоклеточные")), Fact("Запасное вещество - крахмал")
    )
    def Red_algae(self):
        self.declare(Fact(planet="Красные водоросли"))

    @Rule(
        AND(
            Fact("Низшие"),
            Fact("Многоклеточные"),
            Fact("Запасное вещество - ламинарин"),
        )
    )
    def Brown_algae(self):
        self.declare(Fact(planet="Бурые водоросли"))

    @Rule(
        AND(OR(Fact("Колониальные")), Fact("Многоклеточные")),
        Fact("Запасное вещество - масла"),
        Fact("Низшие"),
        Fact("Запасное вещество - лейкозин"),
        Fact("Панцирь клеток из диоксида кремния"),
    )
    def Diatoms(self):
        self.declare(Fact(planet="Диатомовые водоросли"))

    @Rule(
        AND(OR(Fact("Колониальные"), Fact("Многоклеточные")), Fact("Одноклеточные")),
        Fact("Низшие"),
        Fact("Запасное вещество - крахмал"),
    )
    def Green_algae(self):
        self.declare(Fact(planet="Зеленые водоросли"))

    @Rule(
        AND(
            OR(Fact("Колониальные"), Fact("Многоклеточные")),
            Fact("Низшие"),
            Fact("Запасное вещество - масла"),
            Fact("Запасное вещество - лейкозин"),
        )
    )
    def Yellow_green_algae(self):
        self.declare(Fact(planet="Желто-зеленые водоросли"))

    @Rule(
        OR(
            Fact("Опыление - ветромды"),
            Fact("Опыление - животными"),
            Fact("Опыление - водой"),
            Fact("Опыление - самоопыление"),
        )
    )
    def PollinationAngiosperms(self):
        self.declare(Fact("Опыление покрытосеменных"))

    @Rule(
        AND(
            Fact("Есть цветки"),
            Fact("Есть плоды"),
            Fact("Есть орган, улавливающий пыльцу"),
            Fact("Опыление покрытосеменных"),
        )
    )
    def Angiosperms(self):
        self.declare(Fact("Покрытосеменные"))

    @Rule(AND(Fact("Параллельное"), Fact("Дуговое")))
    def VenationMonocots(self):
        self.declare(Fact("Жилкование листьев однодольных"))

    @Rule(
        AND(
            Fact("Корневая система - мочковатая"),
            Fact("Жилкование листьев однодольных"),
        )
    )
    def Monocots(self):
        self.declare(Fact("Однодольные"))

    @Rule(OR(Fact("Перистое"), Fact("Сетчатое")))
    def VenationDuocots(self):
        self.declare(Fact("Жилкование листьев двудольных"))

    @Rule(
        AND(
            Fact("Корневая система - стержневая"), Fact("Жилкование листьев двудольных")
        )
    )
    def Duocots(self):
        self.declare(Fact("Двудольные"))

    @Rule(
        AND(
            Fact("Плод- боб"),
            OR(Fact("Жизненная форма - Травы"), Fact("Жизненная форма - Кустарники")),
            Fact("Двудольные"),
        )
    )
    def Bean(self):
        self.declare(Fact("Бобовые"))

    @Rule(
        AND(Fact("Бобовые"), Fact("Семена продолговатые"), Fact("Паутинистый рисунок"))
    )
    def Peanut(self):
        self.declare(Fact(planet="Арахис"))

    @Rule(AND(Fact("Бобовые"), Fact("Цветки собраны в виде головки")))
    def Clever(self):
        self.declare(Fact(planet="Клевер"))

    @Rule(AND(Fact("Бобовые"), Fact("Семена сферической формы")))
    def Peas(self):
        self.declare(Fact(planet="Горох"))

    @Rule(AND(Fact("Бобовые"), Fact("Семена отделены перегородками")))
    def Bean(self):
        self.declare(Fact(planet="Фасоль"))

    @Rule(
        AND(
            Fact("Двудольные"),
            OR(
                OR(
                    OR(
                        Fact("Жизненная форма - Травы"),
                        Fact("Жизненная форма - Кустарники"),
                    ),
                    Fact("Жизненная форма - Деревья"),
                ),
                OR(
                    Fact("Плод - многоорешек"),
                    Fact("Плод - костянка"),
                    Fact("Плод - яблоко"),
                ),
            ),
        )
    )
    def Rosaceae(self):
        self.declare(Fact("Розоцветные"))

    @Rule(AND(Fact("Розоцветные"), Fact("Жгучие волоски")))
    def Nettle(self):
        self.declare(Fact(planet="Крапива"))

    @Rule(AND(Fact("Розоцветные"), Fact("Плод - яблоко")))
    def Apple(self):
        self.declare(Fact(planet="Яблоня"))

    @Rule(AND(Fact("Розоцветные"), Fact("Есть шипы")))
    def Raspberries(self):
        self.declare(Fact(planet="Малина"))

    @Rule(
        AND(
            Fact("Розоцветные"),
            Fact("Плод - многоорешек"),
            Fact("Соцветие колос"),
            Fact("Соцветие метелка"),
        )
    )
    def Hemp(self):
        self.declare(Fact(planet="Конопля"))

    @Rule(
        AND(Fact("Плод - семянка"), Fact("Двудольные"), Fact("Жизненная форма - Травы"))
    )
    def Compositae(self):
        self.declare(Fact("Сложноцветные"))

    @Rule(AND(Fact("Сложноцветные"), Fact("Высокий стебель")))
    def Sunflower(self):
        self.declare(Fact(planet="Подсолнечник"))

    @Rule(AND(Fact("Сложноцветные"), Fact("2 вида цветков в корзинке")))
    def Chamomile(self):
        self.declare(Fact(planet="Ромашка"))

    @Rule(AND(Fact("Сложноцветные"), Fact("Почки на уровне почвы")))
    def Aster(self):
        self.declare(Fact(planet="Астра"))

    @Rule(
        AND(Fact("Плод - Стручок"), Fact("Двудольные"), Fact("Жизненная форма - Травы"))
    )
    def Cruciferous(self):
        self.declare(Fact("Крестоцветные"))

    @Rule(
        AND(OR(Fact("Желтые лепестки"), Fact("Белые лепестки")), Fact("Крестоцветные"))
    )
    def Cabbage(self):
        self.declare(Fact(planet="Капуста"))

    @Rule(
        AND(
            OR(
                OR(Fact("Желтые лепестки"), Fact("Белые лепестки")),
                Fact("Пурпурные лепестки"),
            ),
            Fact("Съедобные корни"),
        )
    )
    def Radish(self):
        self.declare(Fact(planet="Редька"))

    @Rule(AND(Fact("Есть волоски"), Fact("Белые лепестки"), Fact("Крестоцветные")))
    def Radish(self):
        self.declare(Fact(planet="Редька"))

    @Rule(AND(Fact("Желтые лепестки"), Fact("Крестоцветные")))
    def Mustard(self):
        self.declare(Fact(planet="Горчица"))

    @Rule(
        AND(
            Fact("Жизненная форма - Травы"),
            Fact("Жизненная форма - Кустарники"),
            Fact("Двудольные"),
            OR(Fact("Плод - Коробочка"), Fact("Плод - Ягода")),
        )
    )
    def Cruciferous(self):
        self.declare(Fact("Пасленовые"))

    @Rule(AND(Fact("Пасленовые"), Fact("Подземные съедобные клубни")))
    def Potato(self):
        self.declare(Fact(planet="Картофель"))

    @Rule(AND(Fact("Пасленовые"), Fact("Многогнездные ягоды")))
    def Tomato(self):
        self.declare(Fact(planet="Томат"))

    @Rule(AND(Fact("Пасленовые"), Fact("Шипы на коробочке")))
    def Datura(self):
        self.declare(Fact(planet="Дурман"))

    @Rule(AND(Fact("Пасленовые"), Fact("Шарообразная крышечка коробочки")))
    def Henbane(self):
        self.declare(Fact(planet="Белена"))

    @Rule(
        AND(
            OR(
                Fact("Плод - Коробочка"),
                Fact("Плод - Ягода"),
                Fact("Жизненная форма - Травы"),
                Fact("Однодольные"),
            )
        )
    )
    def Lilyy(self):
        self.declare(Fact("Лилейные"))

    @Rule(AND(Fact("Лилейные"), Fact("Широкие лепестки"), Fact("6 лепестков")))
    def Tulip(self):
        self.declare(Fact(planet="Тюльпан"))

    @Rule(AND(Fact("Лилейные"), Fact("Лепестки загнуты"), Fact("6 лепестков")))
    def Lily(self):
        self.declare(Fact(planet="Лилия"))

    @Rule(
        AND(
            Fact("Плод - Зерновка"),
            Fact("Жизненная форма - Травы"),
            Fact("Однодольные"),
        )
    )
    def Cereals(self):
        self.declare(Fact("Злаковые"))

    @Rule(AND(Fact("Злаковые"), Fact("Быстро растёт")))
    def Bamboo(self):
        self.declare(Fact(planet="Бамбук"))

    @Rule(AND(Fact("Злаковые"), Fact("Соцветие колос")))
    def Wheat(self):
        self.declare(Fact(planet="Пшеница"))

    @Rule(AND(Fact("Злаковые"), Fact("Съедобный початок")))
    def Corn(self):
        self.declare(Fact(planet="Кукуруза"))

    @Rule(AND(Fact("Злаковые"), Fact("Соцветие метёлка")))
    def Feather(self):
        self.declare(Fact(planet="Ковыль"))

    # Вывод
    @Rule(Fact(planet=MATCH.planet))
    def print_planet(self, planet):
        print("Это растение - ", planet)
```
